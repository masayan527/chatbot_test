```python
# N回行った場合の確率（二項分布のグラフ作成）
def plotBinomial(N,p):
    xs = []
    ys = []
    for x in range(N+1):
        xs.append(x)
        ys.append(binom.pmf(x,N,p))
    fig = plt.figure(figsize=(8,5))
    graph = fig.add_subplot(1,1,1)
    graph.bar(xs,ys)
```

```python
# 確率計算
def calcProbability(k,N,p):
    pro = sum(binom.pmf(k,N,p))*100
    return pro
```

```python
# 10回引いて、確率10%
plotBinomial(10,0.1)
```
